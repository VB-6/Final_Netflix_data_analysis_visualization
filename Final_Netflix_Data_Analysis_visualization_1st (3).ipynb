```python
#changing the datatype of date added from object to datetime format

df['date_added']=df['date_added'].astype('datetime64[ns]')
```

<b><h6>Count of Movies , TV Show</h6></b>

```python
type_counts=df['type'].value_counts()
type_counts
```

<b><h6>Top 10 Directors</h6></b>

```python
#top 10 directors 
top_10_director=df['director'].value_counts().head(10)
print(top_10_director)
print("--------------------------------------------------------------------")

# here is not given is high so filtering
filtered=df[df['director']!= 'Not Given']
top_5_director=filtered['director'].value_counts().head(5)
top_5_director
```

<b><h6>Month,Year(Feature Engineering)</h6></b>

```python
#monthly data
df['month']=df['date_added'].dt.month
```

```python
#yearly data
df['year']=df['date_added'].dt.year
```

<b><h6>Monthly release Movies,TV Show</h6></b>

```python

monthly_added_movie=df[df['type']== 'Movie']
Monthly_release_movies=monthly_added_movie.groupby('month').size().reset_index(name='count')
print("monthly_added_movie")
print(Monthly_release_movies)


#monthly release Tv Show
monthly_release_Tv_show=df[df['type']=='TV Show']
monthly_release_tv_show=monthly_release_Tv_show.groupby('month').size().reset_index(name='count')
print("monthly_release_Tv_show")
print(monthly_release_tv_show)
```

<b><h6>count of content added duration Wise(Top 5)</h6></b>


```python
top5_duration_added=df['duration'].value_counts().head(5)
top5_duration_added
```

<b><h6>yearly release Movies,TV Show</h6></b>

```python

yearly_release_movies=df[df['type']=='Movie']
movies_by_year = yearly_release_movies.groupby('year').size().reset_index(name='count')
print("Yearly Added Movies")
print(movies_by_year)

print("-----------------------------------------------------------------------------------")
yearly_release_tv_show=df[df['type']=='TV Show']
tv_shows_by_year= yearly_release_tv_show.groupby('year').size().reset_index(name='count')
print("yearly_release_tv_show")
print(tv_shows_by_year)
```

<b><h6>Top 10 Ratings</h6></b>

<b><h6>Top 10 countries which added content most</h6></b>

```python
top_10_countries=df['country'].value_counts().head(10)
top_10_countries
```

<b><h6>Top 10 Movie Genres</h6></b>

```python
top10_Movie_genres=df[df['type']=='Movie']['listed_in'].value_counts().head(10)
top10_Movie_genres
```

<b><h6>Top 10 TV Show Genres</h6></b>

```python
top10_tv_show_genres=df[df['type']=='TV Show']['listed_in'].value_counts().head(10)
top10_tv_show_genres
```

<b><h1>Exploratory Data Analyis</h1></b>

<b><h3>Count of Movies and TV Show</h3></b>

```python
plt.figure(figsize=(4,4))
sns.barplot(x=type_counts.index , y=type_counts.values, palette='magma')
plt.xlabel("Type")
plt.ylabel("Counts")
plt.title("Count of Movie and TV Shows")
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.title("Distribution of Movies and TV Show")
count_data=sns.countplot(x='type',data=df,hue='type',palette='magma')
plt.grid(False)
for i in count_data.containers:
    count_data.bar_label(i)
plt.savefig('Distribution_of_Movies_and_TV_Show.jpg')
plt.show()
```

<b><h3>Top 5 Directors</h3></b>

```python
plt.figure(figsize=(8,3))
ax=sns.barplot(x=top_5_director.index , y=top_5_director.values , palette='viridis')
plt.xlabel('Directors')
plt.ylabel('Count of Content Added')
plt.title('Top 5 Directors')
plt.xticks(rotation=45)
plt.grid(False)
for i in ax.containers:
    ax.bar_label(i)
plt.savefig('Top_5_Directors.jpg')
plt.show()
```

<b><h3>Top 10 Rating</h3></b>

```python
plt.figure(figsize=(12,6))
ax=sns.barplot(x=top_10_ratings.index , y=top_10_ratings.values , palette='magma')
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Top 10 Rating")
plt.xticks(rotation=45)
plt.grid(False)
for i in ax.containers:
    ax.bar_label(i)
plt.savefig('Top_10_Rating.jpg')
plt.show
```

<b><h3>Rating Comparison Between TV Shows and Movies</h3></b>

```python
plt.figure(figsize=(12,6))
sns.countplot(x='rating',hue='type',data=df,palette='viridis')
plt.title("Rating Comparison Between TV Shows and Movies")
plt.savefig('Rating_Comparison.jpg')
plt.grid(False)
plt.show()
```

<b><h3>Content Added Over Years</h3></b>

```python
plt.figure(figsize=(8, 4))
cnt=sns.countplot(x='year',hue='year',data=df, palette='viridis')
plt.title('Content Added Over Time(Years)')
plt.xlabel('Year')
plt.grid(False)
plt.ylabel('Count')
plt.xticks(rotation=45)
for i in cnt.containers:
    cnt.bar_label(i)
plt.savefig('content_added_over_years.jpg')
plt.show()
```

<b><h3>Movie and TV show comparison(Content Added Over Years)</h3></b>

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='year', y='count', data=movies_by_year, marker='o', color='b',s=100,label='Movie')
sns.scatterplot(x='year' , y='count', data=tv_shows_by_year , marker='*' , color='r',s=100, label='TV Show')
plt.title("Movie and TV show comparison(Content Added Over Years)")
plt.grid(False)
plt.savefig('Movie_TV_Comparison_yearly.jpg')
plt.show()
```

<b><h3>Movie and TV show comparison(Content Added Over Months)</h3></b>

```python
plt.figure(figsize=(8,4))
plt.title("Movie and TV show comparison(Content Added Over Months)")
sns.lineplot(x='month' , y='count' , data=Monthly_release_movies  , marker='o',label='Movies' ,color='g')
sns.lineplot(x='month' , y='count', data=monthly_release_tv_show, marker='*' , label='TV Show' , color='b')
plt.grid(False)
plt.savefig('Movie_TV_Comparison_monthly.jpg')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.title("count of content added duration Wise")
cnt=sns.barplot(x=top5_duration_added.index , y=top5_duration_added.values , palette='magma')
plt.grid(False)
for i in cnt.containers:
    cnt.bar_label(i)
plt.savefig('Duration.jpg')
plt.show()

```
